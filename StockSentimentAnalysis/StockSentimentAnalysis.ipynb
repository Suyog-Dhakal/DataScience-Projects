```python
df4 = df3.copy()
df4.head(1)
```

```python
df4 = df4.drop('Year',axis=1)
df4.head(1)
```

```python
# train = df4[df4['Date'] < '20150101']
# test = df4[df4['Date'] > '20141231']
```

```python
# remocing punctuations
data = df4.iloc[:,2:27]# selecting from 2nd index column to 27th index columns
data.replace("[^a-zA-Z]"," ", regex=True,inplace=True) #replacing all the character except a-z and A-Z by empty space " "
```

```python
#renaming column names for ease of access
list1 = [i for i in range(25)]
new_Index = [str(i) for i in list1]
data.columns = new_Index
data.head(5)
```

```python
#converting headlines to lowercase
for index in new_Index:
    data[index] = data[index].str.lower()
data.head(1)
```

```python
# for every index, appending the string to make a one single paragraph
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```
