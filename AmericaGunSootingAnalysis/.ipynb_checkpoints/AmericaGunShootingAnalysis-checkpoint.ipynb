```python
df2 = df1.copy()
df2['Date'] = df1['Date'].astype('datetime64[ns]')
```

```python
df2.head()
```

```python
df2.drop(columns='Description',axis='columns',inplace=True)
```

```python
df2.head()
```

```python
df2['Year'] = df2['Date'].dt.year
```

```python
df2.head()
```

## EDA

```python
deaths_per_state=df2.groupby(["State"]).sum()['Dead']
deaths_per_state.plot(kind='bar',figsize=(20,7),title='Deaths per State',grid=True)
```

```python
deaths_per_year = df2.groupby('Year').sum()['Dead']
```

```python
deaths_per_year.plot(kind='bar',figsize=(20,7),title='Deaths over the years',grid=True)
```

```python
injuries_per_state = df2.groupby('State').sum()['Injured']
injuries_per_state.plot(kind='bar',figsize=(20,7),title='Injuries per state',grid=True)
```

```python
illinois_row = df2[df2['State'] == 'Illinois'].groupby('State').sum()['Injured']
```

```python
illinois_row
```

```python
injuries_per_year = df2.groupby('Year').sum()['Injured']
injuries_per_year.plot(kind='bar',figsize=(20,7),title='Injuries over the years',grid=True)
```

```python
victims_per_state = df2.groupby('State').sum()['Total']
victims_per_state.plot(kind='bar',figsize=(20,7),title='total victims per state',grid=True)
```

```python
victims_per_year = df2.groupby('Year').sum()['Total']
victims_per_year.plot(kind='bar',figsize=(20,7),title='total victims over the years',grid=True)
```

```python
top_10_dangerous_cities = df2.groupby('City').sum()['Total'][:10].sort_values(ascending=False)
top_10_dangerous_cities.plot(kind='bar',ylabel='Numbers of victims')
```
