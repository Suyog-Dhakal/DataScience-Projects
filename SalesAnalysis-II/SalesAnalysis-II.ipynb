## What is the overall sales trend?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
#getting month year from the dataset
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df['month_year']
```

```python
df_trend = df.groupby('month_year').sum()['sales'].reset_index()
```

```python
df_trend
```

```python
plt.figure(figsize=(15,8))
plt.plot(df_trend['month_year'], df_trend['sales'])
plt.xticks(rotation='vertical', size=8)
plt.xlabel("month-year")
plt.ylabel("Sales number")
plt.show()
```

## Which are the top 10 products by sales?

```python
df.head(20)
```

```python
top_products = df.groupby("product_name").sum()['sales'].reset_index()
```

```python
top_products = top_products.sort_values('sales', ascending=False)
top_products
```

```python
#top 10 products by sales
top_products[:10]
```

## Which are the most selling products?


```python
top_selling_products = df.groupby('product_name').sum()['quantity'].reset_index()
top_selling_products
```

```python
top_selling_products.sort_values('quantity', ascending=False)
top_selling_products[:10]
```
