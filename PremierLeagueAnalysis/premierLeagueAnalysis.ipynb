## Creating predictors for machine learning

```python
# converting 'venue' columns value into numeric value
#converting string into categories and categories into number
#.cat.codes coverts the result into numbers
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches.head()
```

```python
#new column venue_code is added 
# 0 = away , 1 = home
```

```python
#creating unique code for each opponent team in new column
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches.head()
```

```python
#making new 'hour' column for the hour time by splitting value of 'time' column
matches["hour"] = matches["time"].str.replace(":.+","",regex=True).astype("int")
```

```python
matches.head()
```

```python
#creating new column for assigning unique code for each day of the week 
#eg: 0 = monday, 1 = tuesday, 2 = wednesday, ...
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches.head()
```

```python
#target that we are going to predict is whether our team won or not
matches["target"] = (matches["result"] == "W")
```

```python
matches.head()
```

```python
''' 
if result is "W" then true
else false

converting the bool value into number
true = 1, false = 0
'''
matches["target"] = (matches["result"] == "W").astype("int")
```

## Creating initial machine learning model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#here 50 is the number of decision trees
# 10 is the no of sample in the leafnode of the decision tree before splitting
rf = RandomForestClassifier(n_estimators=50,min_samples_split=10,random_state=1)
```

```python
#spliting for training and testing
#training data always should be of past time, testing data should be newer than training
#we can only predict future using the past

train = matches[matches["date"] < '2022-01-01']
test = matches[matches["date"] > '2022-01-01']
```

```python
#list of the predictors columns we created
predictors = ["venue_code","opp_code","hour","day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])#making prediction
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test["target"],preds)
acc
```

```python
#visualinzing in which situation the accuracy was high and low
combined = pd.DataFrame(dict(actual=test["target"],prediction=preds))
```

```python
pd.crosstab(index = combined["actual"], columns = combined["prediction"])
```

```python
#above matrix is confusion matrix
#331 times were correct prediction for draw or lose
#90  times were correct prediction for win
```

```python
#revising accuracy metrics of above 'accuracy_score'
from sklearn.metrics import precision_score
```

```python
#this metrics gives output of : when we predicted a win , what percentage of the time did the team actually win
precision_score(test["target"],preds)
```

## improving precision with rolling averages

```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester United Stats")
```

```python
group
```

```python
#rolling average of 3 weeks to predict the next result
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed="left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
```

```python
cols = ["gf","ga","sh","sot","dist","fk","pk","pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
#here above the index is randomly repeated
#so assigning unique value
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

## retraining the model

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual = test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"],preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
